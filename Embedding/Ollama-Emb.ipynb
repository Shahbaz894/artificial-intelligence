```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader('attention.pdf')
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_documents=text_splitter.split_documents(docs)
final_documents
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents=text_splitter.split_documents(docs)
final_documents
```

```python
print(len(final_documents))
```

```python
embeddings=(OllamaEmbeddings(model='gemma:2b'))
```

```python
first_10_documents = final_documents[:10]
```

```python
r1=embeddings.embed_documents(first_10_documents)
```

```python
r1
```

```python
# Function to process documents in batches
def process_in_batches(documents, batch_size=5):
    results = []
    for i in range(0, len(documents), batch_size):
        batch = documents[i:i + batch_size]
        batch_results = embeddings.embed_documents(batch)
        results.extend(batch_results)
    return results

# Generate embeddings for the first 10 documents in batches
r1 = process_in_batches(first_10_documents, batch_size=5)
```
